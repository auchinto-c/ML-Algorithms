# Logistic Regression

In this example we are trying to train a Logistic Regression model to predict whether a patient may or may not have breast cancer, depending on a set of features.

We are using the `load_breast_cancer` dataset available in `sklearn` package.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split, GridSearchCV
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = load_breast_cancer()
```

```python
data = pd.DataFrame(df['data'], columns = df['feature_names'])
data.head()
```

```python
data['Target'] = df['target']
data.head()
```

```python
## Independent features
X = data.iloc[:,:-1]
X.head()
```

```python
# Dependent feature
Y = pd.DataFrame(data.iloc[:, -1])
Y.head()
```

```python
## Other approach

"""
df = load_breast_cancer()

X = pd.DataFrame(df['data'], columns = df['feature_names'])
Y = pd.DataFrame(df['target'], columns = ['Target'])
"""
```

```python
Y['Target'].value_counts()
```

```python
## Train Test Split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.33, random_state=42)
```

```python
params = [{'C': [1, 5, 10]}, {'max_iter': [100, 150]}]
```

```python
model_1 = LogisticRegression(C = 100, max_iter = 100)
```

```python
model = GridSearchCV(model_1, param_grid = params, scoring = 'f1', cv = 5)
```

```python
model.fit(X_train, Y_train)
```

```python
model.best_params_
```

```python
model.best_score_
```

```python
Y_pred = model.predict(X_test)
Y_pred
```

#### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
confusion_matrix(Y_test, Y_pred)
```

```python
accuracy_score(Y_test, Y_pred)
```

```python
print(classification_report(Y_test, Y_pred))
```
